```python
import numpy as np
import matplotlib.pyplot as plt

print("Graphing Calculator")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Graph a quadratic equation")

choice = input("Choose an option (1-5): ")

if choice in ["1", "2", "3", "4"]:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == "1":
        print("Answer:", num1 + num2)
    elif choice == "2":
        print("Answer:", num1 - num2)
    elif choice == "3":
        print("Answer:", num1 * num2)
    elif choice == "4":
        if num2 == 0:
            print("You can't divide by zero.")
        else:
            print("Answer:", num1 / num2)

elif choice == "5":
    print("Quadratic format: y = ax^2 + bx + c")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))

    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"y = {a}x² + {b}x + {c}")
    plt.axhline(0)
    plt.axvline(0)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Quadratic Graph")
    plt.legend()
    plt.grid(True)
    plt.show()

else:
    print("Invalid choice.")
```
